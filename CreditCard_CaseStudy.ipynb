```python
for col in ol_columns:
    q1 = feature_df[col].quantile(0.25)
    q3 = feature_df[col].quantile(0.75)
    IQR = q3 - q1
    
    upper = q3 + 1.5 * IQR
    lower = q1 - 1.5 * IQR
    
    
    feature_df[col] = feature_df[col].clip(lower=lower, upper=upper)
```

```python
# Total Credit Card Spend (Last 3 months)
df['total_cc_spend'] = df['cc_cons_apr'] + df['cc_cons_may'] + df['cc_cons_jun']
df["total_cc_spend"].head(2)
```

```python
# 2. Average Monthly CC Spend
df['avg_cc_spend'] = df['total_cc_spend'] / 3
df['avg_cc_spend'].head(2)
```

```python
# 3. CC Spend Growth (Month-over-Month)
df['cc_spend_growth_may'] = df['cc_cons_may'] - df['cc_cons_apr']
df['cc_spend_growth_jun'] = df['cc_cons_jun'] - df['cc_cons_may']
```

```python
# 4. Total Debit Card Spend
df['total_dc_spend'] = df['dc_cons_apr'] + df['dc_cons_may'] + df['dc_cons_jun']
```

```python
# 5. CC vs DC Spend Ratio 
df['cc_dc_ratio'] = df['total_cc_spend'] / (df['total_dc_spend'] +1 )
```

```python
# 6. Total Transactions (Last 3 months)
df['total_transactions'] = (df['cc_count_apr'] + df['cc_count_may'] + df['cc_count_jun'] +
                            df['dc_count_apr'] + df['dc_count_may'] + df['dc_count_jun'])
```

```python
# 7. Loan Flags
df['active_loans'] = df['personal_loan_active'] + df['vehicle_loan_active']
df['closed_loans'] = df['personal_loan_closed'] + df['vehicle_loan_closed']
df['total_loans'] = df['active_loans'] + df['closed_loans']
```

```python
# 8. Total Investment
df['total_investment'] = (df['investment_1'] + df['investment_2'] + 
                          df['investment_3'] + df['investment_4'])
```

```python
# 9. EMI Active Flag
df['emi_active_flag'] = df['emi_active'].apply(lambda x: 1 if x > 0 else 0)
```

```python
np.sort(df["age"].unique())
```

```python
# 10. Age Groups
df['age_group'] = pd.cut(df['age'], bins=[0, 25, 35, 45, 60, 100], 
                         labels=['Young', 'Adult', 'Middle', 'Senior', 'Elder'])
```

#### Why Log Transformation Worked So Well Here

* The target (cc_cons) was highly skewed.
* Log transformation normalized the distribution.
* It reduced the disproportionate influence of high spenders.
* It made the error metric (RMSPE) more stable and meaningful.
* The model could now learn relative patterns instead of being dominated by outliers.

`Result: Massive improvement in RMSPE (from 36% → 3.4%).`

#### Why np.log1p() and not np.log() ?
* Zero values: In your data, some customers may have cc_cons = 0 (no spend in 3 months).
* np.log(0) = -inf → This breaks the model (NaN or error).
* np.log1p(0) = 0 → Safe and clean.
